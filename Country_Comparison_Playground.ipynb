```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/omkarkorde27/dataviz/refs/heads/main/joined_HPMD.csv")
```

```python
data.head()
```

```python
lowerGDP_countries = data['Log GDP Per Capita'].quantile(0.25)
lowermiddle_highermiddleGDP_countries = data['Log GDP Per Capita'].median()
higherGDP_countries = data['Log GDP Per Capita'].quantile(0.75)
```

```python
lowerGDP = data[
    (data['Log GDP Per Capita'] <= lowerGDP_countries)
]
countries_with_lowGDP = lowerGDP['Country Name']

lowermiddleGDP = data[
    (data['Log GDP Per Capita'] >= lowerGDP_countries) &
    (data['Log GDP Per Capita'] <= lowermiddle_highermiddleGDP_countries )
]
countries_with_lowermiddleGDP = lowermiddleGDP['Country Name']

highermiddleGDP = data[
    (data['Log GDP Per Capita'] >= lowermiddle_highermiddleGDP_countries) &
    (data['Log GDP Per Capita'] <= higherGDP_countries)

]
countries_with_highermiddleGDP = highermiddleGDP['Country Name']

higherGDP = data[
    (data['Log GDP Per Capita'] >= higherGDP_countries)
]
countries_with_highGDP = higherGDP['Country Name']

print(countries_with_lowGDP, countries_with_lowermiddleGDP, countries_with_highermiddleGDP, countries_with_highGDP)
```

```python
def get_gdp_category(country):
  if country in countries_with_lowGDP.values:
    return 'Low '
  elif country in countries_with_lowermiddleGDP.values:
    return 'lower-middle'
  elif country in countries_with_highermiddleGDP.values:
    return 'higher-middle'
  else:
      return "High"
```

```python
import plotly.graph_objects as go

print("Available countries:")
print(data['Country Name'].unique())
country1 = input("Enter the first country for comparison: ")
country2 = input("Enter the second country for comparison: ")

years = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
print("Available years:")
print(years)
selected_year = int(input("Enter the year you want to compare: "))

if country1 not in data['Country Name'].unique():
    raise ValueError(f"{country1} is not a valid country. Please select from the list.")
if country2 not in data['Country Name'].unique():
    raise ValueError(f"{country2} is not a valid country. Please select from the list.")
if selected_year not in years:
    raise ValueError(f"{selected_year} is not a valid year. Please select from the list.")

country1_data = data[(data['Country Name'] == country1) & (data['Year'] == selected_year)]
country2_data = data[(data['Country Name'] == country2) & (data['Year'] == selected_year)]

if country1_data.empty or country2_data.empty:
    raise ValueError(f"No data available for {country1} or {country2} in {selected_year}.")

country1_category = get_gdp_category(country1)
country2_category = get_gdp_category(country2)

metrics = [
    'Log GDP Per Capita', 'Life Ladder', 'Schizophrenia',
    'Anxiety disorders', 'Depression', 'Bipolar disorders',
    'Eating disorders', 'Alcohol use disorders', 'Drug use disorders'
]
ranges = {metric: (data[metric].min(), data[metric].max()) for metric in metrics}

values_country1 = country1_data[metrics].values[0]
values_country2 = country2_data[metrics].values[0]
annotated_categories = [f"{metric}\n({ranges[metric][0]:.2f} - {ranges[metric][1]:.2f})" for metric in metrics]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=metrics,
    y=values_country1,
    name=f"{country1} ({country1_category})",
    hovertemplate=(
        f"Country: {country1}<br>"
        "Metric: %{x}<br>"
        "Value: %{y}<extra></extra>"
    ),
    marker=dict(color='#CC79A7')
))

fig.add_trace(go.Bar(
    x=metrics,
    y=values_country2,
    name=f"{country2} ({country2_category})",
    hovertemplate=(
        f"Country: {country2}<br>"
        "Metric: %{x}<br>"
        "Value: %{y}<extra></extra>"
    ),
    marker=dict(color='#009E73')
))

fig.update_layout(
    title=f"Comparison of Metrics ({country1} vs. {country2}) - {selected_year}",
    xaxis=dict(
        title="Metrics",
        tickvals=list(range(len(metrics))),
        ticktext=annotated_categories,
    ),
    yaxis_title="Values",
    barmode='group',
    template="plotly_white",
    legend_title="Countries"
)

fig.write_html("comparison_chart_Greece_vs_Uzbekistan.html")
print("Visualization saved to comparison_chart.html")

fig.show()
```

This visualization compares Greece and Uzbekistan on a range of metrics, including GDP per capita, life ladder scores, and various mental health disorders, for a specific year selected by the user. The chart uses a grouped bar chart format, enabling side-by-side comparisons of these metrics.                                                                
                                                                        Key Features:

1. **Group Bar Chart**: Displays values for multiple metrics side by side, with one bar representing each country for each metric.
Categories (metrics) are shown on the x-axis, and their corresponding values are on the y-axis.

2. **Colorblind-Friendly Design**: Two colors were chosen from a colorblind-friendly palette:

-> Purple (#CC79A7) for Greece.

-> Green (#009E73) for Uzbekistan.

3. Hover Information:
Provides interactive tooltips that display:
The metric name.
The country name.
The exact value for the metric.  

Methods Used:

1. Data Processing:
Filtered and validated user inputs (country names and year).
Extracted metric values for the selected countries from the dataset.

2. Plotly Bar Chart:
Created using the **plotly.graph_objects.Bar** trace.
Added two traces (one for each country), with hover templates for interactivity.

The bar chart compares Anxiety Disorders, Log GDP per Capita, and Life Ladder (Happiness Score) between Uzbekistan and Greece for the year 2017. The key insights are:

1. Log GDP Per Capita: Greece(**10.26**) has a significantly higher GDP per capita compared to Uzbekistan(**8.83**). This indicates greater economic prosperity in Greece.

2. Life Ladder (Happiness Score):Uzbekistan's(**6.42**) happiness score is higher than Greece's(**5.14**), even though Uzbekistan has a lower GDP per capita.
This suggests that higher economic wealth does not guarantee greater happiness.

3. Anxiety Disorders: Anxiety disorder prevalence is much higher in Greece(**5.79**) compared to Uzbekistan(**2.56**). This highlights that mental health issues, such as anxiety, are not necessarily tied to economic prosperity and may be influenced by other societal or cultural factors.

Key Takeaway:
The chart reinforces the idea that economic wealth (as represented by GDP) is not the sole determinant of happiness or mental health outcomes. Countries like Uzbekistan, despite lower GDP, may have stronger societal or cultural factors contributing to higher happiness and lower anxiety. This highlights the need to focus on well-being metrics beyond GDP.
